```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading the data
old_df = pd.read_csv("rainfall_data.csv")
old_df.head()
```

```python
#Check for null values in the dataset
old_df.isna().sum()
```

```python
# Create a new DataFrame to store the result
new_data = {
    'Date': ['Latitude', 'Longitude'],
    'Vavuniya': [8.75, 80.5],
    'Anuradhapura': [8.35, 80.38],
    'Maha Illuppallama': [8.12, 80.47]
}

# Extract and format the date and rainfall data
for index, row in old_df.iterrows():
    year = row['Year']
    month = row['Month']
    station_name = row['station_name']
    
    for day in range(1, 32):
        day_str = str(day)
        if day_str in row and not pd.isna(row[day_str]):  # Ensure the day column exists and is not NaN
            date = f'{year:04d}{month:02d}{day:02d}'
            if date not in new_data['Date']:
                new_data['Date'].append(date)
                new_data['Vavuniya'].append(None)
                new_data['Anuradhapura'].append(None)
                new_data['Maha Illuppallama'].append(None)
            
            date_index = new_data['Date'].index(date)
            if station_name == 'VAVUNIYA':
                new_data['Vavuniya'][date_index] = row[day_str]
                #print(f'Adding {row[day_str]} to Vavuniya on {date}')
            elif station_name == 'ANURADHAPURA':
                new_data['Anuradhapura'][date_index] = row[day_str]
                #print(f'Adding {row[day_str]} to Anuradhapura on {date}')
            elif station_name == 'MAHA ILLUPPALLAMA':
                new_data['Maha Illuppallama'][date_index] = row[day_str]
                #print(f'Adding {row[day_str]} to Maha Illuppallama on {date}')

# Convert the dictionary to a DataFrame
new_df = pd.DataFrame(new_data)

# Write the new DataFrame to a CSV file
output_file = 'new_dataset.csv'
new_df.to_csv(output_file, index=False)

# Display the preview of the new DataFrame
print(new_df.head(10))
```

```python
# Create a new DataFrame to store the result
new_data = {
    'Date': [],
    'Vavuniya': [],
    'Anuradhapura': [],
    'Maha Illuppallama': []
}

# Extract and format the date and rainfall data
for index, row in old_df.iterrows():
    year = row['Year']
    month = row['Month']
    station_name = row['station_name']
    
    for day in range(1, 32):
        day_str = str(day)
        if day_str in row and not pd.isna(row[day_str]):  # Ensure the day column exists and is not NaN
            date = f'{year:04d}{month:02d}{day:02d}'
            if date not in new_data['Date']:
                new_data['Date'].append(date)
                new_data['Vavuniya'].append(None)
                new_data['Anuradhapura'].append(None)
                new_data['Maha Illuppallama'].append(None)
            
            date_index = new_data['Date'].index(date)
            if station_name == 'VAVUNIYA':
                new_data['Vavuniya'][date_index] = row[day_str]
                #print(f'Adding {row[day_str]} to Vavuniya on {date}')
            elif station_name == 'ANURADHAPURA':
                new_data['Anuradhapura'][date_index] = row[day_str]
                #print(f'Adding {row[day_str]} to Anuradhapura on {date}')
            elif station_name == 'MAHA ILLUPPALLAMA':
                new_data['Maha Illuppallama'][date_index] = row[day_str]
                #print(f'Adding {row[day_str]} to Maha Illuppallama on {date}')

# Convert the dictionary to a DataFrame
df = pd.DataFrame(new_data)

# Write the new DataFrame to a CSV file
output_file = 'transformed_rainfall_data.csv'
df.to_csv(output_file, index=False)

# Display the preview of the new DataFrame
print(df.head(10))
```

```python
# Generate summary statistics
print("Summary Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Plot the distribution of rainfall data for each station
stations = ['Vavuniya', 'Anuradhapura', 'Maha Illuppallama']
for station in stations:
    plt.figure(figsize=(12, 6))
    plt.hist(df[station].dropna(), bins=50, alpha=0.7, label=station)
    plt.title(f'Distribution of Rainfall in {station}')
    plt.xlabel('Rainfall')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

```python
# Correlation analysis
correlation_matrix = df[stations].corr()
plt.figure(figsize=(12, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Seasonal Analysis
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
```

```python
# Average monthly rainfall for each station
monthly_avg = df.groupby('Month')[stations].mean()
monthly_avg.plot(figsize=(12, 6))
plt.title('Average Monthly Rainfall')
plt.xlabel('Month')
plt.ylabel('Average Rainfall')
plt.legend()
plt.show()
```

```python
# Boxplot to see the distribution of rainfall across different months
for station in stations:
    plt.figure(figsize=(12, 6))
    df.boxplot(column=station, by='Month', grid=False)
    plt.title(f'Monthly Rainfall Distribution in {station}')
    plt.xlabel('Month')
    plt.ylabel('Rainfall')
    plt.show()
```
